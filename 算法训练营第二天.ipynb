```python
# 977. 有序数组的平方
# 双指针，两个序列，都是从大到小，比较两个序列中的最大值，更大的放在后面
# 如果跳到一个序列中了，更大的还是放在后面
class Solution:
    def sortedSquares(self, nums: List[int]) -> List[int]:
        for i, num in enumerate(nums):
            nums[i] = num * num
        length = len(nums)
        l = 0
        r = length - 1
        ans = [0] * length
        i = length - 1
        while l <= r:
            if nums[l] > nums[r]:
                ans[i] = nums[l]
                l += 1
            else:
                ans[i] = nums[r]
                r -= 1
            i -= 1
        return ans
        
```

```python
# 209. 长度最小的子数组
# 滑动窗口，双指针，右指针不断向右移动，直到满足条件，左指针向右移动，直到不满足条件
class Solution:
    def minSubArrayLen(self, target: int, nums: List[int]) -> int:
        sum = 0
        ans = float('inf')
        k = 0
        for i, num in enumerate(nums):
            sum += num
            while sum >= target:
                ans = min(ans, i - k + 1)
                sum -= nums[k]
                k += 1
        return ans if ans != float('inf') else 0
            
```
