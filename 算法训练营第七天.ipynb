```python
## 四数之和
## 先计算前两个数的和，并记录这个和出现的次数，存入字典，再计算后两个数的和，看是否在字典中
class Solution:
    def fourSumCount(self, nums1: List[int], nums2: List[int], nums3: List[int], nums4: List[int]) -> int:
        cnt = 0
        mp = {}
        for a in nums1:
            for b in nums2:
                mp[a + b] = mp.get(a + b, 0) + 1
        for c in nums3:
            for d in nums4:
                cnt += mp.get(-(c + d), 0)
        return cnt
```

```python
## 383. 赎金信
## 统计a字符串中每个字符出现的次数，然后遍历b字符串，抵消a中出现的字符，如果出现负数，说明b中出现的字符在a中没有，返回False

class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        mp = {}
        for c in magazine:
            mp[c] = mp.get(c, 0) + 1
        for c in ransomNote:
            mp[c] = mp.get(c, 0) - 1
            if mp[c] < 0:
                return False
        return True
```

```python
# 202. 三数之和
class Solution:
    def threeSum(self, nums: List[int]) -> List[List[int]]:
        n = len(nums)
        nums.sort()
        ans = []
        for i in range(0, len(nums) - 2):
            if i > 0 and nums[i] == nums[i - 1]:
                continue
            target = -nums[i]
            l = i + 1
            r = n - 1
            # 去重逻辑，只有当[l, r]这个区间里没有数的时候才结束
            while l < r:
                if nums[l] + nums[r] < target:
                    l += 1
                elif nums[l] + nums[r] > target:
                    r -= 1
                else:
                    ans.append([nums[i], nums[l], nums[r]])
                    # l < r 保证了遍历不会越界，因为l后面一定会有一个数
                    while l < r and nums[l] == nums[l + 1]:
                        l += 1
                    while l < r and nums[r - 1] == nums[r]:
                        r -= 1
                    l += 1
                    r -= 1
        return ans
        
            
```

```python
# 18. 四数之和
class Solution:
    def fourSum(self, nums: List[int], target: int) -> List[List[int]]:
        nums.sort()
        ans = []
        for i in range(0, len(nums) - 3):
            if i > 0 and nums[i] == nums[i - 1]: # 这里需要去重
                continue
            if nums[i] > target and nums[i] > 0:
                continue
            for j in range(i + 1, len(nums) - 2):
                if j > i + 1 and nums[j] == nums[j - 1]:
                    continue
                if nums[i] + nums[j] > target and target > 0:
                    continue
                l = j + 1
                r = len(nums) - 1
                t = target - nums[i] - nums[j]
                while l < r:
                    if nums[l] + nums[r] == t:
                        ans.append([nums[i], nums[j], nums[l], nums[r]])
                        while l < r and nums[l] == nums[l + 1]:
                            l += 1
                        while l < r  and nums[r] == nums[r - 1]:
                            r -= 1
                        l += 1
                        r -= 1
                    elif nums[l] + nums[r] < t:
                        l += 1
                    else:
                        r -= 1
        return ans
```
