```python
# 232. 用栈实现队列
# 关键点：栈1的数入栈2的时候，保证栈2是空的，这样栈2的顺序就是队列的顺序
class MyQueue:
    def __init__(self):
        self.stk1 = []
        self.stk2 = []
        self.size = 0
    def push(self, x: int) -> None:
        self.stk1.append(x)
        self.size += 1

    def pop(self) -> int:
        if self.size > 0:
            if not self.stk2:
                while self.stk1:
                    self.stk2.append(self.stk1.pop())
            self.size -= 1
            return self.stk2.pop()

    def peek(self) -> int:
        if not self.stk2:
            while self.stk1:
                self.stk2.append(self.stk1.pop())
        return self.stk2[-1]



    def empty(self) -> bool:
        return self.size <= 0



# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```

```python
# 用队列实现栈
# 关键点：出栈的时候，将除最后一个元素外的其他元素都出队，然后再入队，最后再pop队头元素
from collections import deque

class MyStack:

    def __init__(self):
        self.q = deque()


    def push(self, x: int) -> None:
        self.q.append(x)


    def pop(self) -> int:
        for _ in range(len(self.q) - 1):
            self.q.append(self.q.pop())
        return self.q.pop()

    def top(self) -> int:
        return self.q[-1]


    def empty(self) -> bool:
        return len(self.q) <= 0



# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()
```
