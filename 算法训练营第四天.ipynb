### 面试题 02.07. 链表相交


```python
class Solution:
    def getIntersectionNode(self, headA: ListNode, headB: ListNode) -> ListNode:
        curA = headA
        curB = headB
        while curA != curB:
            curA = curA.next if curA is not None else headB
            curB = curB.next if curB is not None else headA
        return curA
```

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def detectCycle(self, head: Optional[ListNode]) -> Optional[ListNode]:
        fast = head
        slow = head
        while fast is not None and fast.next is not None:
            fast = fast.next.next
            slow = slow.next
            if fast == slow:
                break
        if fast is None or fast.next is None:
            return None
        while head != fast:
            head = head.next
            fast = fast.next
        return fast
```

### 24. 两两交换链表中的节点


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        dummyHead = ListNode(-1, head)
        cur = dummyHead
        while cur.next is not None and cur.next.next is not None:
            first = cur.next
            second = cur.next.next
            cur.next = second
            first.next = second.next
            second.next = first
            cur = cur.next.next
        return dummyHead.next
            
```

```python
# 对比上一份代码可知
# first 和 second 是两个指针，分别指向要交换的两个节点
# 但是second是基于first不为空的情况下才能存在的
# 所以在while循环中，每次要判断first是否为空
# 而cur是指向first的前一个节点，在while给的前提下，cur永远可以正常更新
# 所以尽量不要在循环外面定义指针的next，因为循环内部的指针可能为空，会报错
class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head is None:
            return head
        dummyHead = ListNode(-1, head)
        cur = dummyHead
        first = head
        second = head.next
        while second is not None: # 当前循环条件不能保证更新后的first不为空
            cur.next = second
            first.next = second.next
            second.next = first
            cur = cur.next.next
            first = first.next
            if first is None:
                break
            second = first.next
        return dummyHead.next
            
```

### 19. 删除链表的倒数第 N 个结点


```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dummyHead = ListNode(-1, head)
        fast = dummyHead
        for _ in range(n + 1):
            fast = fast.next
        slow = dummyHead
        while fast is not None:
            fast = fast.next
            slow = slow.next
        slow.next = slow.next.next
        return dummyHead.next
        
```
